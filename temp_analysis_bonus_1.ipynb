```python
import pandas as pd
from datetime import datetime as dt
from scipy.stats import ttest_ind
```

```python
#read the csv file
df = pd.read_csv('hawaii_measurements.csv')
df
```

```python
#convert the date column format from string to datetime
df["date"] =  pd.to_datetime(df["date"], format="%Y/%m/%d")
df.dtypes
```

```python
#set the date column as the dataframe index
date_index= df.set_index('date')
date_index
```

```python
#drop the date column
drop_date = date_index.reset_index(drop=True)
drop_date
```

```python
#filter the data for all temperatures in June
june = df.loc[df['date'].dt.month == 6]

#filter the data for all temperatures in December
december = df.loc[df['date'].dt.month==12]
```

```python
#get the average temperature for June
avg_june = june['tobs'].mean()
avg_june
```

```python
#get the average temperature for December
avg_dec= december['tobs'].mean()
avg_dec
```

```python
#utilized an independent t-test to compare temperatures in June and December
#the t-test cannot conclude that a significant diference exists in the results
res = ttest_ind(june['tobs'],december['tobs'])
res
```
